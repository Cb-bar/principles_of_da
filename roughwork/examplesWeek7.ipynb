## Boxplots

```python
# Plotting.
import matplotlib.pyplot as plt

# Numerical Arrays.
import numpy as np
```

## Continuous v. Discrete Data

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]

plt.bar(fruits, counts)
```

## Creating Data

Create a random number generator, with a deterministic seed.

https://numpy.org/doc/stable/reference/random/generator.html

```python
# Random number generator
rng = np.random.default_rng(12345)
```

Generate some random normal data

https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.normal.html

```python
# Group 1: Mean = 75, Std Dev = 10, Size = 30
group_1 = rng.normal(75,10,30)

# Show
print(group_1)
```

## Bar Chart v. Histograms

```python
# List of unique values and their counts.
unique, counts = np.unique(group_1, return_counts=True)

# Show
unique, counts
```

```python
plt.bar(unique, counts)
```

```python
# Histogram
plt.hist(group_1, edgecolor='black')
```

## More Groups

```python
# Group 2: Mean = 65, Std Dev = 12, Size = 30
group_2 = rng.normal(65,12,30)

# Show
print(group_2)
```

```python
# Group 3: Mean = 85, Std Dev = 8, Size = 30
group_3=rng.normal (85, 8, 30)

# Show.
print (group_3)
```

```python
# Histograms
plt.hist(group_1, edgecolor='black', label='Group 1', alpha=0.5)
plt.hist(group_2, edgecolor='black', label='Group 2', alpha=0.5)
plt.hist(group_3, edgecolor='black', label='Group 3', alpha=0.5)

plt.legend()
```

## Organise Data

https://numpy.org/doc/stable/reference/generated/numpy.array.html

```python
# Create a numpy array.
data = np.array([group_1, group_2, group_3])

# Show.
print(data)
```

## Descriptive Statistics

Mean: https://numpy.org/doc/2.2/reference/generated/numpy.mean.html

Median: https://numpy.org/doc/2.2/reference/generated/numpy.median.html

Standard Deviation: https://numpy.org/doc/2.2/reference/generated/numpy.std.html

```python
# Mean
np.mean(group_1)
```

```python
# Means
np.mean(data, axis=1)
```

```python
# Means.
print(f'Means: {np.mean(data, axis=1)}')

# Standard Deviation
print(f'Standard Deviations: {np.std(data ,axis=1)}')

# Medians
print(f'Medians: {np.median(data, axis=1)}')

# Maxima
print(f'Maxima: {np.max(data, axis=1)}')

# Minima
print(f'Minima: {np.min(data, axis=1)}')
```

## Quartiles

https://numpy.org/doc/stable/reference/generated/numpy.quantile.html

https://www.codecademy.com/learn/dscp-summarizing-quantitative-data/modules/dscp-quartiles-quantiles-and-interquartile-range/cheatsheet

```python
# Q1
np.quantile(data, 0.25, axis=1)
```

```python
# Q3
np.quantile(data, 0.75, axis=1)
```

```python
# Quantiles
np.quantile(data, [0.25, 0.5, 0.75], axis=1)
```
