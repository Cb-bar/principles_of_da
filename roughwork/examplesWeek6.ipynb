## matplotlib

```python
# Plots
import matplotlib.pyplot as plt

# Numerical arrays
import numpy as np
```

## Plotting Functions

```python
# X Values
x = np.arange(0.0, 10.0, 0.1)

# Show
x
```

```python
# X Values
x = np.linspace(0.0, 10.0, 101)

# Show
x
```

```python
# Y Values
y = x**2

# Show
y
```

```python
# Plot x vs. y.
plt.plot(x, y, "g.", label='$y=x^2$');

plt.legend();
plt.xlabel('x');
plt.ylabel('$x^2$');
plt.title('A Lovely Plot');
```

## Comparing Plots

```python
# Domain of x values.
x = np.linspace(0.0, 10.0, 101)

# x squared.
y1 = x**2

# 2 to the power of x.
y2 = 2**x
```

```python
# Plot x vs. y1.
plt.plot(x, y1, label='$x^2$')

# Plot x vs. y2.
plt.plot(x, y2, label='$2^x$')

# Add a legend.
plt.legend()

# Add labels
plt.xlabel("x")
plt.ylabel("y")

# Add a title.
plt.title('Two Lovely Plots');
```

## Integers

```python
# Domain x values.
x = np.linspace(0, 10, 11, dtype=int)

# Show X 
x
```

```python
x.dtype
```

```python
# x cubed plus 2x squared plus 1.
y = x**3 + 2 * x**2 + 1

# Show y
y
```

```python
# Type of y
y.dtype
```

```python
# Plot x vs. y
plt.plot(x, y, '.', label='$x^3 + 2*x^2 + 1$')

# Add a legend.
plt.legend()

# Add labels
plt.xlabel("x")
plt.ylabel("y")

# Add a title.
plt.title('Plotting Ints');
```
